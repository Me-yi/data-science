## `.loc()`

Label based

```python
brics_russia=brics_csv.loc['RU']
print(brics_russia)
print(type(brics_russia))
```

Remember to convert, not to `Series`, but to a `DataFrame`

```python
brics_russia=brics_csv.loc[['RU']]
print(brics_russia)
print(type(brics_russia))
```

This can be expanded to other indexed labels

```python
brics_asia=brics_csv.loc[['RU','IN','CH']]
print(brics_asia)
```

### Column specific selection

```python
brics_asia=brics_csv.loc[['RU','IN','CH'], ['country','capital']]
print(brics_asia)
```

```python
brics_country_capital=brics_csv.loc[:, ['country','capital']]
print(brics_country_capital)
```

## `.iloc()`

Integer position-based

```python
# Results are exactly the same
brics_loc = brics_csv.loc['RU']
brics_iloc = brics_csv.iloc[1]
print(f"""{brics_loc}\n\n{brics_iloc}""")
```

We can do the same for column and rows

```python
brics_country_capital = brics_csv.loc[:, ['country','capital']]
brics_country_capital_iloc = brics_csv.iloc[:, [0,1]]
print(f"""{brics_country_capital}\n\n{brics_country_capital_iloc}""")
```
