## Compare NumPy Arrays

```python
my_house = np.array([18.0, 20.0, 10.75, 9.50])
your_house = np.array([14.0, 24.0, 14.25, 9.0])
```

```python
# my_house greater than or equal to 18
print(my_house >=18)

# my_house less than your_house
print(my_house < your_house)
```

## Boolean operators with NumPy

These boolean operators exist, as the built-up syntax does not allow for `and` or `or` Python boolean operators to be used

```python
# my_house greater than 18.5 or smaller than 10
print(np.logical_or(my_house > 18.5, my_house < 10))

# Both my_house and your_house smaller than 11
print(np.logical_and(my_house < 11, your_house < 11))
```

## Filtering Panda Dataframes

### Creating `Series`

```python
# Extract drives_right column as Series: dr
dr = cars['drives_right']

# Use dr to subset cars: sel
sel = cars[dr]

# Print sel
print(sel)
```

## Comparison Operators

```python
# Create car_maniac: observations that have a cars_per_cap over 500
many_cars = cars['cars_per_cap'] > 500
car_maniac = cars[many_cars]

# Print car_maniac
print(car_maniac)
```

## NumPy Boolean Operator Variants

```python
# Create medium: observations with cars_per_cap between 100 and 500
cpc = cars['cars_per_cap']
between = np.logical_and(cpc > 100, cpc < 500)
medium = cars[between]

# Print medium
print(medium)
```

## Boolean Variants In-Depth

```python
print(f'    np.logical_or([True, True], [True, True]) -> {np.logical_or([True, True], [True, True])}')
print(f'   np.logical_or([True, False], [True, True]) -> {np.logical_or([True, False], [True, True])}')
print(f'  np.logical_or([False, False], [True, True]) -> {np.logical_or([False, False], [True, True])}')
print(f' np.logical_or([False, False], [False, True]) -> {np.logical_or([False, False], [False, True])}')
print(f' np.logical_or([False, False], [False, True]) -> {np.logical_or([True, False], [False, False])}')
print(f'np.logical_or([False, False], [False, False]) -> {np.logical_or([False, False], [False, False])}')
```

The following arrays are boolean compared like so `[False, False, True]` and `[True, False, True]`

The index are boolean compared together.

`False or True` is `True`

`False or False` is `False`

`True or True` is `True`
