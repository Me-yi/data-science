## Required Imports

```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
```

## Required Data

```python
brics_uri = '/content/drive/My Drive/Colab Notebooks/data/learning/brics.csv'
brics = pd.read_csv(brics_uri, index_col = 0)
print(brics)
```

## for, first try

```python
for table_headings in brics:
  print(table_headings)
```

## `iterrows` method

```python
for label, row in brics.iterrows():
  print(label)
  print(row)
```

### Print out capitals

```python
for label, row in brics.iterrows():
  print(f"{label}: {row["capital"]}")
```

## Adding an extra column

```python
for label, row in brics.iterrows():
  brics.loc[label, "name_length"] = len(row["country"])
print(brics)
```

### Using `apply` DataFrame column

More efficient way to add on a  DataFrame column; use `apply`.

```python
brics["name_length"] = brics["country"].apply(len)
print(brics)
```
