## Google Cloud

### Retrieving CSV files from Google Cloud Storage (GCS)

To access files from Google Cloud, you typically store them in a GCS bucket. You will need to authenticate your session to access your Google Cloud resources.

```python
from google.colab import auth
auth.authenticate_user()

print('Authenticated successfully')
```

```python
import google.auth

try:
    credentials, project = google.auth.default()
    print(f"Authenticated as: {credentials.service_account_email}")
    print(f"Default Project ID: {project}")
except Exception as e:
    print(f"Authentication failed or not yet performed: {e}")
```

Once authenticated, you can use the `google-cloud-storage` library to download the blob (file) or use `pandas` to read it directly if you have the `gcsfs` library installed.

```python
import pandas as pd

# Replace these with your actual bucket and file name
bucket_name = 'your-bucket-name'
file_name = 'data.csv'
gcs_path = f'gs://{bucket_name}/{file_name}'

# Project ID is required for billing/access
project_id = 'your-project-id'

try:
    df_gcs = pd.read_csv(gcs_path, storage_options={'project': project_id})
    print(f"Successfully loaded {gcs_path}")
    display(df_gcs.head())
except Exception as e:
    print(f"Ensure you have 'gcsfs' installed and the path is correct: {e}")
```

```python
# Assuming you have a CSV file named 'my_data.csv' in the root of your Google Drive
# You might need to adjust the path based on your file's actual location in Drive.
# For example, if it's in a folder called 'Colab Notebooks', the path would be '/content/drive/MyDrive/Colab Notebooks/my_data.csv'

drive_file_path = '/content/drive/MyDrive/my_data.csv'

try:
    df_from_drive = pd.read_csv(drive_file_path)
    print(f"Successfully loaded data from {drive_file_path}")
    display(df_from_drive.head())
except FileNotFoundError:
    print(f"Error: The file '{drive_file_path}' was not found. Please ensure the file exists in your Google Drive and the path is correct.")
except Exception as e:
    print(f"An error occurred while reading the CSV file: {e}")
```

## Environment session storage

Using the session storage is option, to quickly include a simple file to store.

Although this isn't really the recommended way. Once the session is finished - the runtime close and data once retrieved is gone.

### Uploading CSV files to Colab

To upload a CSV file to your current Colab session, you can use the file icon on the left sidebar (Folder icon) -> 'Upload to session storage'. Once uploaded, the file will be available in the `/content/` directory and can be accessed by its filename directly. Be aware that these files are temporary and will be deleted once the session ends.
